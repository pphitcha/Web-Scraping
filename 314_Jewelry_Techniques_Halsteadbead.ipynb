```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
i_page = 1
title_list = []
detail_list = []

while i_page <=11:
    data = requests.get(f'https://www.halsteadbead.com/articles/search/jewelry%20techniques/p/{i_page}')
    data.encoding = 'utf8'
    soup = bs4.BeautifulSoup(data.text)
    for item in soup.find_all('div', class_='post-listing'):
        title_list.append(item.find('h3').find('a').text.strip())
        detail_list.append(item.find('div', class_='post-listing-text').text.strip())
        
    print('Complete page number: ',i_page)
    i_page += 1
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
print(soup)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.DataFrame([title_list,detail_list]).transpose()
df.columns = ['title','detail']
df
```

```python
df['detail_new'] =df['detail'].str.split('\r\n\t\t\t\t\t\t\n\n\r\n\t\t\t\t\t\t').str[1]
df
```

```python
df1 = df.drop(['detail'], axis=1)
df1
```

```python
df2 =  df1.replace({'\’':'\'','\s\s':' '}, regex = True)
df2
```
