```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
i_page = 1
title_list = []
detail_list = []

while i_page <= 3:
    data = requests.get(f'https://www.langantiques.com/university/jewelry-techniques/?_page={i_page}')
    data.encoding = 'utf8'
    soup = bs4.BeautifulSoup(data.text)
    for item in soup.find_all('div', class_='pt-cv-ifield'):
        title_list.append(item.find('h4').find('a').text.strip())
        detail_list.append(item.find('div', class_='pt-cv-content').text.strip())
        
    print('Complete page number: ',i_page)
    i_page += 1
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
print(soup)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.DataFrame([title_list,detail_list]).transpose()
df.columns = ['title','detail']
df
```
