```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
i_page = 1
title_list = []
detail_list = []

while i_page <= 34:
    data = requests.get(f'https://www.gemsociety.org/search/Jewelry_Technique/page/{i_page}/')
    data.encoding = 'utf8'
    soup = bs4.BeautifulSoup(data.text)
    for item in soup.find_all('div', class_='col-md-4'):
        title_list.append(item.find('h5').find('a').text.strip())
        detail_list.append(item.find('div', class_='article-preview__excerpt').text.strip())
        
    print('Complete page number: ',i_page)
    i_page += 1
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
print(soup)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.DataFrame([title_list,detail_list]).transpose()
df.columns = ['title','detail']
df
```

```python
df1 =  df.replace({'\“':'\"','\”':'\"','\’':'\'','…':'',' – ':'-','\©':' ','é':'e',' — ':'-','[^\x00-\x7F]+':''}, regex = True)
df1
```
