```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.jaumelabro.com/?s=jewelry+techniques"
```

```python
res = requests.get(url)
res.encoding = 'utf8'
print(res)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
print(soup)
```

```python
title_list = []
detail_list = []

for item in soup.find_all('article'):
        title_list.append(item.find('h2').find('a').text.strip())
        detail_list.append(item.find('p').text.strip())
```

```python
df = pd.DataFrame([title_list,detail_list]).transpose()
df.columns = ['title','detail']
df
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df1 =  df.replace({'\’':'\'','\[…\]':'',' – ':'-','[^\x00-\x7F]+':''}, regex = True)
df1
```
