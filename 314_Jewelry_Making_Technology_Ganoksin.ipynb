```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
i_page = 1
title_list = []
detail_list = []

while i_page <=16:
    data = requests.get(f'https://www.ganoksin.com/page/{i_page}/?s=Jewelry-Making+Technology')
    data.encoding = 'utf8'
    soup = bs4.BeautifulSoup(data.text)
    for item in soup.find_all('article', class_='article-preview'):
        title_list.append(item.find('h5', class_='article-preview__title').text.strip())
        detail_list.append(item.find('div', class_='article-preview__excerpt').text.strip())
        
    print('Complete page number: ',i_page)
    i_page += 1
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
print(soup)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.DataFrame([title_list,detail_list]).transpose()
df.columns = ['title','detail']
df
```
